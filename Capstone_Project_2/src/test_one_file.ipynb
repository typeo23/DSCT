```python
import gensim
import pandas as pd
```

```python
data_dir = '../data/'
filename = 's2-corpus-00.gz'
```

```python
reader = pd.read_json(data_dir+filename, lines=True, chunksize=1e5)
```

```python
df = next(reader)
```

```python
from gensim.models.word2vec import Word2Vec
model= Word2Vec.load('doc2vec_one_file2.model')
```
